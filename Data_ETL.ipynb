```python
import pandas as pd
from sqlalchemy import create_engine
```

### EDUCATION

```python
# Store Education Attainment CSV into a Data Frame
csv_file_1 = "Data_Sets/Education.csv"
education_raw_df = pd.read_csv(csv_file_1)
education_raw_df.head()
```

```python
# Replace spaces in column headers with underscore, and delete cuotes
education_raw_df.columns = education_raw_df.columns.str.replace(' ', '_')
education_raw_df.columns = education_raw_df.columns.str.replace("'", '')
education_raw_df.head()
```

```python
# Create new Data Frame with select columns
education_raw_df = education_raw_df[[
    'Federal_Information_Processing_Standard_(FIPS)_Code',
    'State',
    'Area_name',
    'Percent_of_adults_with_less_than_a_high_school_diploma,_2016-20',
    'Percent_of_adults_with_a_high_school_diploma_only,_2016-20',
    'Percent_of_adults_completing_some_college_or_associates_degree,_2016-20',
    'Percent_of_adults_with_a_bachelors_degree_or_higher_2015-19'
    ]].copy()
education_raw_df.head()
```

```python
#Rename columns with shorter names
education_raw_df.columns = [
    'FIPS',
    'State',
    'County',
    'Less_than_HS',
    'HS',
    'Some_college',
    'BS_or_higher'
    ]
education_raw_df.head()
```
