### LIFE EXPECTANCY

```python
# Store Life Expectancy CSV into a Data Frame
csv_file_3 = "Data_Sets/Life_Expectancy.csv"
life_raw_df = pd.read_csv(csv_file_3)
life_raw_df.head()
```

```python
# Replace spaces in column headers with underscore
life_raw_df.columns = life_raw_df.columns.str.replace(' ', '_')
life_raw_df.head()
```

```python
# Create new Data Frame with select columns
life_raw_df = life_raw_df[[
    'Location',
    'FIPS',
    'Life_expectancy,_2014*',
    ]].copy()
life_raw_df.head()
```

```python
#Rename columns with shorter names
life_raw_df.columns = [
    'County',
    'FIPS',
    'Life_expectancy'
    ]
life_raw_df#.head()
```

```python
#Delete "County" from the County column
life_raw_df['County']=life_raw_df['County'].str.replace(' County','')
life_raw_df.head()
```

```python
#Split "County" in two, one for County and another for State, using the comma as delimiter
life_raw_df[['County2','State']] = life_raw_df.County.str.split(", ",expand=True)
life_raw_df = life_raw_df.drop('County', axis=1)
life_raw_df.columns = [
    'FIPS',
    'Life_expectancy',
    'County',
    'State'
    ]
life_raw_df.head()
```

```python
#Eliminate male and female life expectancy numbers
life_raw_df[['Life',"aux"]] = life_raw_df.Life_expectancy.str.split(', ',expand=True)
life_raw_df[['Life2',"aux2"]] = life_raw_df.Life.str.split(' ',expand=True)
life_raw_df.head()
```

```python
#Eliminate excess columns 
life_raw_df = life_raw_df.drop('Life_expectancy', axis=1)
life_raw_df = life_raw_df.drop('Life', axis=1)
life_raw_df = life_raw_df.drop('aux', axis=1)
life_raw_df = life_raw_df.drop('aux2', axis=1)
life_raw_df.head()
```

```python
#Renaming column and reordering
life_raw_df.rename(columns={'Life2':'Life_expectancy','State':'State_name'}, inplace=True)
life_raw_df = life_raw_df[['FIPS','State_name','County','Life_expectancy']]
life_raw_df.head()
```

```python
#Joining Life Expectancy and State Codes data frames to use state abbreviatoins
life_raw_df = pd.merge(life_raw_df, 
                      state_codes_df, 
                      on ='State_name', 
                      how ='inner')
life_raw_df.head()
```

```python
#Reorder columns
life_raw_df = life_raw_df[['FIPS','State','State_name','County','Life_expectancy']]
life_raw_df.head()
```

```python
#Convert FIPS column to string
life_raw_df['FIPS']= life_raw_df['FIPS'].map(str)
life_raw_df.head()
```

```python
#Split FIPS in two using "." as delimiter and deleting temporary column
life_raw_df[['FIPS','temp']] = life_raw_df.FIPS.str.split(".",expand=True)
life_raw_df = life_raw_df.drop('temp', axis=1)
life_raw_df.head()
```

## MERGING ALL THE DATA INTO ONE TABLE

```python
#List of all the dataframes (comment out to show select ones)
state_codes_df.head()
education_raw_df.head()
GDP_raw_df.head()
life_raw_df.head()
poverty_raw_df.head()
```

```python
#Join all 4 tables
human_development_df = education_raw_df.merge(GDP_raw_df,on='FIPS').merge(poverty_raw_df,on='FIPS').merge(life_raw_df,on='FIPS')
human_development_df.head()
```

```python
#Delete duplicate columns
human_development_df = human_development_df.loc[:,~human_development_df.columns.duplicated()]
human_development_df = human_development_df.drop('County_y', axis=1)
human_development_df = human_development_df.drop('State_y', axis=1)
human_development_df = human_development_df.drop('State_name_y', axis=1)
human_development_df.head()
```

```python
# Reorder columns
human_development_df = human_development_df[['FIPS','State_x','State_name_x','County_x','Less_than_HS','HS','Some_college','BS_or_higher','GDP_2015','Poverty_pct','Life_expectancy']]
human_development_df.head()
```

```python
#Rename columns
human_development_df.columns = [
    'FIPS',
    'State',
    'State_name',
    'County',
    'Less_than_HS',
    'HS',
    'Some_college',
    'BS_or_higher',
    'GDP_2015',
    'Poverty_pct',
    'Life_expectancy'
    ]
human_development_df.head()
```

```python
#Convert FIPS from string to numeric
human_development_df.FIPS=human_development_df.FIPS.astype(int)
human_development_df.dtypes
```

```python
#Set FIPS as index
human_development_df.set_index('FIPS')
```

```python
#Load dataframe into csv
human_development_df.to_csv('human_development_df')
```
